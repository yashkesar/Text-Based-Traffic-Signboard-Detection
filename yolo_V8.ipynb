```python
import os
from PIL import Image, ImageOps

def process_images_with_tesseract(input_folder):
    pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/bin/tesseract'  # Update this path if necessary

    for file_name in os.listdir(input_folder):
        if file_name.endswith('.jpg') or file_name.endswith('.png'):
            # Load image
            input_path = os.path.join(input_folder, file_name)
            image = Image.open(input_path)

            # Get DPI or use a default value
            dpi = image.info.get('dpi', (72, 72))

            # Calculate new size based on 300 DPI and maintain aspect ratio
            target_dpi = 300
            new_width = int((image.width * target_dpi) / dpi[0])
            new_height = int((image.height * target_dpi) / dpi[1])

            # Create new image with calculated size and paste original image into the center
            new_image = Image.new('RGB', (new_width, new_height), color='black')
            paste_x = (new_width - image.width) // 2
            paste_y = (new_height - image.height) // 2
            new_image.paste(image, (paste_x, paste_y))

            # Run Tesseract OCR
            text = pytesseract.image_to_string(new_image)

            # Print output in Colab
            print(f"OCR output for {file_name}:")
            print(text)
            print('-' * 80)
```
